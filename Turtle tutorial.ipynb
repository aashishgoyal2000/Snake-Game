```python
import turtle as robot

def rectangle(horizontal, vertical, color):
    robot.pendown()
    robot.pensize(1)
    robot.color(color)
    robot.begin_fill()
    for counter in range(1,3):
        robot.forward(horizontal)
        robot.right(90)
        robot.forward(vertical)
        robot.right(90)
    robot.end_fill()
    robot.penup()

robot.shape('turtle')
robot.setheading(0)
# robot.forward(80)

robot.penup()
robot.speed('slow')
robot.bgcolor('Dodger blue')

#feet
robot.goto(-100, -150)
rectangle(50, 20, 'blue')
robot.goto(-30, -150)
rectangle(50, 20, 'blue')

# legs
robot.goto(-25, -50)
rectangle(15, 100, 'grey')
robot.goto(-55, -50)
rectangle(-15, 100, 'grey')

# body
robot.goto(-90, 100)
rectangle(100, 150, 'red')

# arms
robot.goto(-150, 70)
rectangle(60, 15, 'grey')
robot.goto(-150, 110)
rectangle(15, 40, 'grey')

robot.goto(10, 70)
rectangle(60, 15, 'grey')
robot.goto(55, 110)
rectangle(15, 40, 'grey')

# neck
robot.goto(-50, 120)
rectangle(15, 20, 'grey')

# head
robot.goto(-85, 170)
rectangle(80, 50, 'red')

# eyes
robot.goto(-60, 160)
rectangle(30, 10, 'white')
robot.goto(-55, 155)
rectangle(5, 5, 'black')
robot.goto(-40, 155)
rectangle(5, 5, 'black')

# mouth
robot.goto(-65, 135)
rectangle(40, 5, 'black')

robot.hideturtle()
```
