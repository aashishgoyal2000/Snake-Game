```python
import random
import turtle as t

t.bgcolor("yellow")

# crete caterpillar
caterpillar = t.Turtle()
caterpillar.shape('square')
caterpillar.color('red')
caterpillar.speed(0)
caterpillar.penup()
caterpillar.hideturtle()

# create leaf 
leaf = t.Turtle()
leaf_shape = ((0,0), (14, 2), (18, 6), (20, 20), (6, 18), (2, 14))
t.register_shape('leaf', leaf_shape)
leaf.shape('leaf')
leaf.color('green')
leaf.penup()
leaf.hideturtle()
leaf.speed(0)

# crete text visible before game start
game_started = False
text_turtle = t.Turtle()
text_turtle.write('Press SPACE to state', align='center', font=('Arial', 16, 'bold'))
text_turtle.hideturtle()

# create score counter
score = t.Turtle()
score.hideturtle()
score.speed(0)

# placeholder functions
def outside_window():
    left = -t.window_width() / 2
    right = t.window_width() / 2
    top = t.window_height() / 2
    bottom = -t.window_height() / 2
    (x, y) = caterpillar.pos()
    outside = x < left or x > right or y < bottom or y > top
    return outside

def game_over():
    caterpillar.color('yellow')
    leaf.color('yellow')
    t.penup()
    t.hideturtle()
    t.write('GAME OVER!', align = 'center', font = ('Arial', 30, 'normal'))

def display_score(turtle_score):
    score.clear()
    score.penup()
    x = (t.window_height() / 2) - 50
    y = (t.window_width() / 2) - 50
    score.setpos(x, y)
    score.write(str(turtle_score), align='right', font=('Arial', 40, 'bold'))

def place_leaf():
    leaf.ht()
    leaf.setx(random.randint(-200, 200))
    leaf.sety(random.randint(-200, 200))
    leaf.st()

segments = []
    
def start_game():
    global game_started
    if game_started:
        return
    
    game_started = True
    
    turtle_score = 0
    text_turtle.clear()
    
    caterpillar_speed = 2
    caterpillar_length = 1
    caterpillar.shapesize(1, caterpillar_length, 1)
    caterpillar.showturtle()
    caterpillar.goto(0, 0)
    display_score(turtle_score)
    place_leaf()
    
    # turtle movement
    while True:
        caterpillar.forward(caterpillar_speed)
        if caterpillar.distance(leaf) < 20:
            place_leaf()    
            caterpillar_speed = caterpillar_speed + 1
            new_segment = t.Turtle()
            new_segment.shape('square')
            new_segment.color('red')
            new_segment.length = 1
            new_segment.speed = caterpillar_speed 
            new_segment.shapesize(1, new_segment.length, 1)
            new_segment.penup()
            new_segment.showturtle()
            
            segments.append(new_segment)            
            turtle_score = turtle_score + 10
            display_score(turtle_score)
        
        for index in range(len(segments) - 1, 0, -1):
            x = segments[index - 1].xcor()
            y = segments[index - 1].ycor()
            segments[index].goto(x, y)
            
        if len(segments) > 0 :
            x = caterpillar.xcor()
            y = caterpillar.ycor()
            segments[0].goto(x, y)
            
        if outside_window():
            if len(segments) <= 0:
                text_turtle.write('Press SPACE to state', align='center', font=('Arial', 16, 'bold'))
                text_turtle.hideturtle()   
                caterpillar.hideturtle()
                game_started = False
                break
            segments.pop(-1)
            
def move_up():
    if caterpillar.heading() == 0 or caterpillar.heading() == 180:
        caterpillar.setheading(90)

def move_down():
    if caterpillar.heading() == 0 or caterpillar.heading() == 180:
        caterpillar.setheading(270)

def move_left():
    if caterpillar.heading() == 90 or caterpillar.heading() == 270:
        caterpillar.setheading(180)

def move_right():
    if caterpillar.heading() == 90 or caterpillar.heading() == 270:
        caterpillar.setheading(0)
    

t.onkey(start_game, 'space')
t.onkey(move_up, 'Up')
t.onkey(move_right, 'Right')
t.onkey(move_down, 'Down')
t.onkey(move_left, 'Left')
t.listen()
t.mainloop()
```
